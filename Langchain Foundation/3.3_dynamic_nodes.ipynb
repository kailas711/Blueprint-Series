### Dynamic Agents

We could have different type of users interacting with our agent , not the same person everytime. Example a user can speak English another one would be speaking French. The user maybe internal or external, so they need access to information or models that other's don't. Building a one-size fit for all agent can be useful for an MVP but for real world deployment, when it comes to **Production Ready Agents**. 

we need an agent that can dynamically change. This is the focus of Dynamic Agents. 

How can we change the prompts, tools and even the models you build your agents with, on the fly. Even multiple times during the conversation. For that we need to build a custom middleware. 




Node style middleware is realy great at hooking into agent and changing it's state. Great for modifying messages before or after we invoke the model. 

**Model Wrap Middle Ware**

The model instance is represented as a model request. This contains information like the system prompt, available tool calls, the state and the foundation model itself. 



We're essentially changing how the model looks like, swapping the model ( like llama with qwen ), changing the system prompt or adapting the system prompt to changing state.
